```python
df_stop_times = pd.read_csv('stop_times.txt')
def parse_time (pd_series, lenght):
    i=0
    while (i<lenght):
        if ((ord(pd_series[i][0]) == ord('2')) and (ord(pd_series[i][1]) > ord('3'))):
            s = list(pd_series[i])
            s[0] = '0'
            s[1] = chr(ord(pd_series[i][1]) - 4)
            pd_series[i] = "".join(s)
        datetime.datetime.strptime(pd_series[i], "%H:%M:%S")
        i = i+1
             
parse_time(df_stop_times['departure_time'],237852)
parse_time(df_stop_times['arrival_time'],237852)
```
